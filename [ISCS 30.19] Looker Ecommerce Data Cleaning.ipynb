```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
# load datasets

df_products = pd.read_csv('products.csv')
df_order_items = pd.read_csv('order_items.csv')
df_inventory = pd.read_csv('inventory_items.csv')
df_orders = pd.read_csv('orders.csv')
df_users = pd.read_csv('users.csv')
```

# **Products**

```python
df_products.head()
```

```python
df_products.info()
```

```python
# drop unused columns
df_products = df_products.drop(columns=['distribution_center_id'])
```

```python
# remove rows with missing product details
df_products = df_products.dropna(subset=['name', 'brand'])
```

```python
# rename id column to product_id
df_products = df_products.rename(columns={'id': 'product_id'})
```

```python
df_products.info()
```

# **Users**

```python
df_users.head()
```

```python
df_users.info()
```

```python
# drop unused columns
df_users = df_users.drop(columns=['first_name',	'last_name',	'email',
                                  'street_address',	'postal_code',	'city', 'state', 'latitude',	'longitude',	'traffic_source',	'created_at'])
```

```python
# rename id column to user_id
df_users = df_users.rename(columns={'id': 'user_id'})
```

```python
df_users.info()
```

# **Orders**

```python
df_orders.head()
```

```python
df_orders.info()
```

```python
# drop unused columns
df_orders = df_orders.drop(columns=['status', 'gender', 'created_at', 'returned_at',	'shipped_at',	'delivered_at',	'num_of_item'])
```

```python
df_orders.info()
```

# **Order Items**

```python
df_order_items.head()
```

```python
df_order_items.info()
```

```python
# drop unused columns
df_order_items = df_order_items.drop(columns=['product_id', 'sale_price', 'user_id', 'shipped_at', 'delivered_at','returned_at'])
```

```python
# remove all orders that were cancelled
df_order_items = df_order_items[df_order_items['status'] != 'Cancelled']
df_order_items =df_order_items.drop(columns='status')
```

```python
# rename id column to order_items_id
df_order_items = df_order_items.rename(columns={'id': 'order_items_id'})

#rename created at to sold at
df_order_items = df_order_items.rename(columns={'created_at': 'sold_at'})
```

```python
df_order_items.info()
```

# **Inventory**

```python
df_inventory.head()
```

```python
df_inventory.info()
```

```python
# drop unused columns
df_inventory = df_inventory.drop(columns=['cost',	'sold_at', 'product_category',	'product_name',	'product_brand',	'product_retail_price',	'product_department',	'product_sku',	'product_distribution_center_id'])
```

```python
# rename id column to inventory_id
df_inventory = df_inventory.rename(columns={'id': 'inventory_item_id'})

# rename created_at to stocked_at
df_inventory = df_inventory.rename(columns={'created_at': 'stocked_at'})
```

```python
df_inventory.info()
```

```python
# Saves the cleaned/transformed DataFrame to CSV file
df_products.to_csv ('products_cleaned.csv', index = False, header=True)
df_order_items.to_csv ('order_items_cleaned.csv', index = False, header=True)
df_inventory.to_csv ('inventory_cleaned.csv', index = False, header=True)
df_orders.to_csv ('orders_cleaned.csv', index = False, header=True)
df_users.to_csv ('users_cleaned.csv', index = False, header=True)


from google.colab import files

files.download('products_cleaned.csv')
files.download('order_items_cleaned.csv')
files.download('inventory_cleaned.csv')
files.download('users_cleaned.csv')
files.download('orders_cleaned.csv')
```
